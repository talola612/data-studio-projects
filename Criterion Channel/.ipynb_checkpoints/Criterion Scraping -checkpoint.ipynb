```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
}
response = requests.get('https://www.criterion.com/shop/browse/list', headers = headers)
doc = BeautifulSoup(response.text)
```

```python
films = doc.find_all('tr')[1:]
year = doc.find_all('td', class_ = 'g-year')
```

```python
rows = []
for film in films:
    row = {}
    row['title'] = film.find(class_ = 'g-title').text.strip()
    row['url'] = film.find(class_ = 'g-title').a['href']
    row['spine'] = film.find(class_ = 'g-spine').text.strip()
    row['img'] = film.find(class_ = 'g-img').img['src']
    row['director'] = film.find(class_ = 'g-director').text.strip()
    row['country'] = film.find(class_ = 'g-country').text.strip().strip(',')

    rows.append(row)
    
rows
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
years = []
for i in range(1, len(films)):
    row = {}
    row['year'] = ([y.text.strip() for y in year][i])
 
    years.append(row)
```

```python
df2 = pd.DataFrame(years)
merge = df2.merge(df, left_index = True, right_index = True)
```

```python
merge.head()
```

```python
merge = merge[['title', 'director', 'country', 'spine', 'url', 'img']]
```

```python
merge
```

```python
merge.to_csv('Criterion_collection.csv', index = False)
```

```python

df = pd.read_csv('Criterion_collection.csv')

film_df = df[df.director.notna()]
film_df.head()

def movie_info(dataset):
    rows = []
    for data in dataset.url:
        headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
}
        response = requests.get(data, headers = headers)
        doc = BeautifulSoup(response.text)
        print('scraping', data)
        row = {}
        row['title'] = doc.h1.text
        row['year'] = doc.find(class_ = 'film-meta-list').find(itemprop = 'datePublished')['content']
        row['description'] = doc.find(class_ = 'product-summary').text.strip()
        row['duration'] = doc.find(class_ = 'film-meta-list').find(itemprop = 'duration').find_parent().text
        try:
            row['color'] = doc.find(class_ = 'film-meta-list').find(itemprop = 'inLanguage').find_previous_sibling().find_previous_sibling(itemprop = False).text
        except:
            pass
        row['aspect_ratio'] = doc.find(class_ = 'film-meta-list').find(itemprop = 'inLanguage').find_previous_sibling().text
        row['lang'] = doc.find(class_ = 'film-meta-list').find(itemprop = 'inLanguage').text.strip()
        
        row['dir_link'] = doc.find(class_ = 'film-meta-list').li.a['href']
        po = doc.findAll(class_ = 'purchase-option')
        for i in range(0, len(po)):
            item_name = doc.findAll(class_ = 'purchase-option')[i].find(class_ = 'item').text
            item_sale_price = doc.findAll(class_ = 'purchase-option')[0].find(class_ = 'item-price').text.strip()
            item_retail_price = doc.findAll(class_ = 'purchase-option')[0].find('span', class_ = 'strike').text
            row[item_name + '_sp'] = item_sale_price
            row[item_name + '_rp'] = item_retail_price
        rows.append(row)
    return pd.DataFrame(rows)
            
    

    
```

```python
info_df = movie_info(film_df)
```
