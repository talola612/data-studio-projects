```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
url = 'http://www.cinemetrics.lv/satltdb.php'
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

# First Table: Camera Movement Data

```python
table = doc.table

output_rows = []
for table_row in table.findAll('tr')[1:]:
    columns = table_row.findAll('td')
    output_row = []
    for column in columns:
        output_row.append(column.text)
    output_rows.append(output_row)

output_rows
```

```python
df = pd.DataFrame.from_records(output_rows)
df.head()
```

```python
df.columns = ['title', 'year', 'director', 'pan', 'tilt', 'pan_w_tilt', 'track', 'tr_w_pan', 'tr_w_pan_n_tilt', 'crane', 'zoom', 'total']
```

```python
df.head()
```

```python
df['title'] = df.title.str.replace('    ', ' ')
```

```python
df['title'] = df.title.str.replace('(.+)\, (.+)', '\\2 \\1')
```

```python
df['director'] = df.director.str.replace('(.+)\, (.+)', '\\2 \\1')
df.head()
```

```python
df.to_csv('Camera_movement.csv', index = False)
```
